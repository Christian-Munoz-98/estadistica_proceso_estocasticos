```python
import numpy as np
import matplotlib.pyplot as plt

x= [2.4, 3.4, 4.6, 3.7, 2.2, 3.3, 4.0, 2.1]
y= [1.33, 2.12, 1.80, 1.65, 2.0, 1.76, 2.11, 1.63]

x_label = 'Cosecha de ma√≠z'
y_label = 'Cosecha de cacahuates'


plt.scatter(x,y)
plt.xlabel(x_label)
plt.ylabel(y_label)
plt.show()

anp = np.array([[2.4, 3.4, 4.6, 3.7, 2.2, 3.3, 4.0, 2.1],
  [1.33, 2.12, 1.80, 1.65, 2.0, 1.76, 2.11, 1.63]])

corrcoef = np.corrcoef(anp)
print(corrcoef) 
```

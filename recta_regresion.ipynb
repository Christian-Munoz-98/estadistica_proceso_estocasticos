```python
import numpy as np 
import matplotlib.pyplot as plt
x = [100, 110, 115, 120, 130, 135, 140, 150, 160, 165, 170, 180, 190, 195,]
x_np = np.array(x) 
y = [45, 51, 53, 54, 61, 63, 66, 70, 74, 76, 78, 85, 89, 91] 
y_np = np.array(y)
media_x = np.mean(x) 
media_y = np.mean(y)
s_x = np.std(x)
s_y = np.std(y)
coef_cor = np.corrcoef(x,y)
print(f"Datasets:\n{x}\n{y}")
print(f"Media de X = {media_x}")
print(f"Media de Y = {media_y}")
print(f"Desviación estandar de X = {s_x}")
print(f"Desviación estandar de Y = {s_y}")
print(f"Coeficiente de correlacion = {coef_cor}")
plt.scatter(x,y)
plt.show()
m =  0.99848449*(np.std(y)/np.std(x))  
b = np.mean(y) - m*np.mean(x) 
plt.plot(x,y,'o') 
plt.plot(x,m*x_np+b) 
plt.show()
```
